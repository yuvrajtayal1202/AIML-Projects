```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model  import LogisticRegression
```

```python
df = pd.read_csv('data.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
df['diagnosis'] = df['diagnosis'].astype(str).str.lower()
df['diagnosis'] = df['diagnosis'].map({'m': 1, 'b': 0})
df = df.drop(columns=['Unnamed: 32', 'id'], errors='ignore')
```

```python
X = df.drop(columns=['diagnosis'])
Y = df['diagnosis']
```

```python
print(X)
print(Y)
```

```python
Standard_Scaler = StandardScaler()
X = Standard_Scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=12)
```

```python
model = LogisticRegression()

model.fit(X_train, Y_train)
```

```python
pred_on_train = model.predict(X_train)
acc_on_train = accuracy_score(Y_train, pred_on_train)
print(acc_on_train)
```

```python
pred_on_test = model.predict(X_test)
acc_on_test = accuracy_score(Y_test, pred_on_test)
print(acc_on_test)
```
