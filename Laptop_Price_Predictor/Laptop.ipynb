## EDA

```python
def info_of_col(col):
    print(f'Unique Values: \n {df[col].unique()} \n')
    print(f'Value Count: \n {df[col].value_counts()} \n')
    df[col].value_counts().plot(kind='bar')
    
def rel_with_price(col):
    print('\n Relation With Price:')
    sns.barplot(x = df[col], y = df['Price'])
    plt.xticks(rotation = 'vertical')
    plt.show()
```

```python
df['Cpu_name']= df['Cpu'].apply(lambda x:x.split()[0:3])
df['Cpu_name']= df['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text 
    else:
        if text.split()[0] == 'Intel':
            return "Other Intel Processor"
        else:
            return 'AMD Processor'
```

```python
df['Cpu_name'].apply(fetch_processor)
```

```python
df['Cpu_brand'] = df['Cpu_name'].apply(fetch_processor)
```

```python
df.drop(columns = ["Cpu", "Cpu_name"], inplace = True)
```

## Operating System

```python
rel_with_price('OpSys')
```

```python
def cat_os(inp):
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'
    else:
        return 'Others/No OS/Linux'
```

```python
df['OS'] = df['OpSys'].apply(cat_os)
```

```python
df['OpSys'].value_counts().plot(kind = 'bar')
```

```python
df.drop(columns = ['OpSys'], inplace = True)
```
